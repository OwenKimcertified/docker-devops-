### element

```python 

services:
  todo-app:
    build:
      context: ./app
    depends_on:
      - todo-database
    environment:
      NODE_ENV: production
    ports:
      - 3000:3000
      - 35729:35729
    develop:
      watch:
        - path: ./app/package.json
          action: rebuild
        - path: ./app
          target: /usr/src/app
          action: sync

  todo-database:
    image: mongo:6
    volumes: 
     - database:/data/db
    ports:
      - 27017:27017

volumes:
  database:

  ```

build: 
  context: ./app => 도커 파일로 빌드하고, 도커 파일의 경로  

depends_on:
=> 의존성 파트

environment:
=> 환경 변수 설정. production 으로 설정하여 개발 단계에서 필요한 것들을 적용 x 

ports:
  - 3000:3000
  - 35729:35729 
=> 포트 설정 app : 3000, mongoDB : 35729

develop:
  watch:
    - path: ./app/package.json
      action: rebuild
    - path: ./app
      target: /usr/src/app
      action: sync

real - time update 를 위한 element

`./app/package.json 에 변경사항이 있을 때 rebuild`

`rebuild 한 ./app/package.json 을 (로컬 파일 시스템)`

`/usr/src/app 에 동기화. (도커 컨테이너 내부)`
